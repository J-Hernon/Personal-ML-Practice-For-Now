Variables

A variable is a placeholder for a different value. They are defined with the "=" character

```python
# An integer is a whole number
int1 = 5 

# A float is a fraction
float1 = 5.5

# A string is a collection of characters. A number can be a string, but you won't be able to add two strings like numbers
str1 = "Cat"
str2 = "56"

# A Boolean is either a true or false statement
tru = True
fal = False

# Use print(type()) to check the type of a certain variable
print(type(int1))
```

Numeric Operators

This will discuss the basic numeric operators used in python, what they mean, and how to use them. They all will return either a float or integer, depending on what the input types were. 

```python
# The + operator is used to add two numbers 
print("The sum is", 5 + 8)

# The - operator is used to subtract two numbers
print("The difference is", 8 - 3)

# The * operator is used to multiply two numbers
print("The product is", 8 * 3)

# The / operator is used to divide two numbers
print("The result is", 8 / 3)

# The ** operator is used to do exponentials two numbers
print("The exponent is", 8 ** 3)

# The % operator is used to get the remainder in a division. 
print("The remainder is", 8 % 3)

# The // operator is for floor division, ie round result down to an integer 
print("The result is", 8 // 5)

# Note adding = to the end of any of these operators will reassign after a value is modified
val1 = 5
val1 += 5
print("After adding 5, it is", val1)
```

Comparison Operators

These are used to compare two datatypes (usually int or floats). The output is a Boolean. These are usually good for conditional statements or loop conditions. 

```python
# The == operator is used to determine if two things are the same
print("Are 5 and 6 the same? ", 5 == 6)

# The != operator is used to determine if two things are NOT the same
print("Are 5 and 6 different? ", 5 != 6)

# The > operator is used to determine if one thing is greater than the other
print("Is 6 greater than 5? ", 6 > 5)

# The < operator is used to determine if one thing is smaller than the other
print("Is 5 greater than 5? ", 5 < 5)

# The >= operator is used to determine if one thing is greater than or equal to the other
print("Is 6 greater than or equal to 5? ", 6 >= 5)

# The <= operator is used to determine if one thing is smaller than or equal to the other
print("Is 5 greater than or equal to 5? ", 5 <= 5)

```

Logical Operators

Logical operators are used to determine if two statements are true. It returns a boolean. 

```python
# The and operator returns True if both things are True
print(5==5 and 6==6)

# The or operator returns True if either thing is True
print(2==3 and 5==5)

# The not operator returns the opposite boolean from the item
print(not 5==5)
```

Identity Operators 

These are used to see if two variables share the same location in memory. It returns a boolean. NOTE: == can return True but is can return False. Probably not used too much. 

```python
# The is operator is used to determine if two variables point to the same location in memory
var1 = 5
var2 = var1 
print(var2 is var1) # True bc point to the same location in memory 

# The is not operator is used to determine if two variables do not point to the same location in memory
var3 = 5
print(var3 is not var1)
```

Membership Operators

These are used to see if something is in a list, tuple, or string. It will return a boolean. 

```python
# The in operator returns True if the thing is in the other item
temp_list = [1,2,3]
print(3 in temp_list)

# The not in operator returns True if the thing is not in the other item
temp_list = [1,2,3]
print(3 not in temp_list)
print(5 not in temp_list)
```

Conditionals

These are used to help sequence code. They will execute a segment if a statement or condition is true, and will pass if false. 

```python
# if is used to execute code if a statement is true. 
grade = 2
if grade < 3: 
    print("FAIL") # EXECUTES HERE
elif grade < 6: 
    print("LOW PASS")
else: 
    print("HIGH PASS")

# elif is used to execute code if an if statement is not initially met, but an other one is. 
# these are good for multiple levels (ie not true or false, but 4-5 possibilites)
grade = 5
if grade < 3: 
    print("FAIL")
elif grade < 6: 
    print("LOW PASS") # First if not met, but first elif is. So Execute here. 
else: 
    print("HIGH PASS")

# else is used to execute code if no conditions are met. 
grade = 10
if grade < 3: 
    print("FAIL")
elif grade < 6: 
    print("LOW PASS")
else: 
    print("HIGH PASS")
```

Loops

These are used to iterate over a segement of code a certain number of times. There are two types: for and while. 

```python
# for loops will execute a given number of times. We define it the number of times it will execute. 
for i in range (5):  # executes over values 0-4
    print('for loop:', i)

# while loops are executed for an unknown number of times. While the given statment is true, the code
# will keep executing. Once it is False, it will stop running. 
stop_val = 6
start_val = 0
while start_val != stop_val: # see if the start value is not equal to the stop val. Since its not, we execute
    print('while loop: ', start_val)
    start_val += 1 # increase the value to the new condition. 
```
